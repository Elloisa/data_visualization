```python
import pandas as pd
#matplotlib is the library we'll be using for visualization
import matplotlib.pyplot as plt
#Seaborn is actually build on top of matplotlib: The two libraries can be 
#seamlesly integrated and work alongside each other.
import seaborn as sns
sns.set()
```

```python
df_used_cars=pd.read_csv("bar_chart_data.csv")
df_used_cars
```

```python
#Shift+Tab to get help or detail
plt.figure(num="Car by brand",
          figsize=(12,6),
          #dpi=500,
          facecolor="white",
           edgecolor="blue",
           frameon=True,
           tight_layout=True
          )
plt.bar(x=df_used_cars["Brand"],
        height=df_used_cars["Cars Listings"],
       #color=["r","g","b","w","y","m","c"] #r:red,g:green,b:blue,linen,honeydew
        color="midnightblue")
plt.xticks(rotation=360, fontsize=13)
plt.yticks(fontsize=13)
plt.title("Cars Listings by Brand", fontsize=16, fontweight="bold")
plt.ylabel("Number of Listings", fontsize=13, labelpad=10)
plt.savefig("Used Car Bar.png")
plt.show()
```

```python
plt.figure(figsize=(9,6),
          facecolor="white")
plt.barh(y=df_used_cars["Brand"], 
        width=df_used_cars["Cars Listings"],
        color="midnightblue")
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)
plt.title("Cars Listing by Brand", fontsize=16, fontweight="bold")
plt.xlabel("Number of Listing", fontsize=13, labelpad=10)
plt.savefig("Used Car hBar.png")
plt.show()
```


# Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_type=pd.read_csv("C://Users/CASA/365/Python/Visualization/data_base/pie_chart_excel.csv")
```

```python
df_fuel_engine_type
```

```python
df_fuel_engine_type.columns.values
```

```python
list(df_fuel_engine_type)
```

```python
#https://www.delftstack.com/howto/python-pandas/split-column-in-python-pandas/
df_fuel_engine_type=df_fuel_engine_type["Engine Fuel Type;Number of Cars"].str.split(pat=";", n=-1,expand=True)
```

```python
df_fuel_engine_type
```

```python
df_fuel_engine_type.columns=["Engine Fuel Type", "Number of Cars"]
df_fuel_engine_type.columns.values
df_fuel_engine_type
```

```python
#seaborn color palette
sns.set_palette("colorblind")
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_fuel_engine_type['Number of Cars'],
       labels=df_fuel_engine_type["Engine Fuel Type"].values,
       autopct= "%.2f%%",
       textprops={"size":"x-large",
                 "fontweight":"bold",
                 "rotation":"40",
                   "color":"w"})
plt.legend()
plt.title("Cars by Engine Fuel Type", fontsize=18, fontweight="bold")
plt.show()
```
